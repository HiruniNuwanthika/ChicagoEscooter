```python
def dailyStartGraph():
    df_list=[]
    for fileID in range (1,8):
        df=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\August\Started\Hourly\hourlyTripCount_Aug 0{} 2022.csv'.format(fileID))
        df.columns = ["hour", "count"]
        df.set_index('hour',inplace=True)
        df_list.append(df)
    df_month = pd.concat(df_list, axis=1).reset_index()
    df_month.columns= ["hour", "1/08","2/08","3/08","4/08","5/08","6/08","7/08"]
    df_month.to_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\August\Started\PerDay\1week.csv'.format(fileID),index=False)
```

```python
dailyStartGraph()
```

```python
def dailyStopGraph():
    df_list=[]
    for fileID in range (1,4):
        df=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\August\Stopped\Hourly\hourlyTripCount_Aug 0{} 2022.csv'.format(fileID))
        df.columns = ["hour", "count"]
        df.set_index('hour',inplace=True)
        df_list.append(df)
    df_month = pd.concat(df_list, axis=1).reset_index()
    df_month.columns= ["hour", "1/08","2/08","3/08"]
    df_month.to_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\August\Stopped\PerDay\1.csv'.format(fileID),index=False)
```
