```python
def GetHourlyTripRecords(day):
    for hour in range(0,24):
        df_list=[]
        for fileID in range(1,5):
            try:
                df =pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\July\{}_{}hour_part{}_final.csv'.format(day,hour,fileID))
            except FileNotFoundError:
                continue
            df_list.append(df)
            all_df=pd.concat(df_list,ignore_index=True)
            try:
                all_df.to_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\July\Hourly\{}_hour{}.csv'.format(day,hour), index=False)
            except UnboundLocalError:
                print ("UnboundLocalError"+str(hour)+str(fileID))
```
