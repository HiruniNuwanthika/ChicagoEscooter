```python
import pandas as pd
import statsmodels.api as sm
```

## all independent variables 

```python
df=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\POI calculations\Method2\POIs\features_selected.csv')
```

```python
Y=df['trip density']
```

```python
X=df[['House/ Apartment perc','Restaurant/ Cafe perc','Shops perc','Leisure/ Recreation','Campus','tram density','bus density','train density','entropy','MXI','walkScore']]
```

```python
X = sm.add_constant(X)
```

```python
ks = sm.OLS(Y, X)
ks_res =ks.fit()
ks_res.summary()
```
