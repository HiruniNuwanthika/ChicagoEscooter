```python
import pandas as pd
```

```python
df=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\Results\July\29_9hour_part1.csv')
```

```python
stopped_list=[]
for i in range(1,16):
    if 'lat_%s' %(i+1) in df.columns:
        stopped=df[(df['lat_%s' %i].isnull()==True) & (df['lat_%s' %(i+1)].isnull()==False)]
        stopped_list.append(stopped)
#print(stopped_list)

stopped_df= pd.DataFrame()
for i in stopped_list:
        stopped_df =  stopped_df.append(i, ignore_index=True)
```
