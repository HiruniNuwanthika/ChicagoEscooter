```python
import pandas as pd
```

```python
df=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\Results\July\29_9hour_part1.csv')
```

```python
stopped_list=[]
for i in range(1,16):
    if 'lat_%s' %(i+1) in df.columns:
        stopped=df[(df['lat_%s' %i].isnull()==True) & (df['lat_%s' %(i+1)].isnull()==False)]
        stopped_list.append(stopped)
#print(stopped_list)

stopped_df= pd.DataFrame()
for i in stopped_list:
        stopped_df =  stopped_df.append(i, ignore_index=True)
```

create list of data frames containing stopped locations.Create a DF of that list

```python
stopped_df
```

```python
stopped_df.to_csv(r'C:\Melbourne_Escooter\RealDataset\Results\July\Stopped\29_9hour_part1.csv',index=False)
```

```python
started_list=[]
for i in range(1,16):
    if 'lat_%s' %(i+1) in df.columns:
        started=df[(df['lat_%s' %i].isnull()==False) & (df['lat_%s' %(i+1)].isnull()==True)]
        started_list.append(started)
#print(stopped_list)

started_df= pd.DataFrame()
for i in started_list:
        started_df =  started_df.append(i, ignore_index=True)
```

```python
started_df
```

```python
started_df.to_csv(r'C:\Melbourne_Escooter\RealDataset\Results\July\Started\29_9hour_part1.csv',index=False)
```

```python
df_start_stop = pd.merge(started_df, stopped_df, how='outer', on = 'bike_id', suffixes=("_start", "_stop"))
```

```python
df_start_stop
```

```python
df_start_stop.to_csv(r'C:\Melbourne_Escooter\RealDataset\Results\July\Start_Stop(outer)\29_9hour_part1.csv', index=False)
```
