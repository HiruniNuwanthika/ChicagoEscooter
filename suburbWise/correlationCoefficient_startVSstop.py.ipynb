```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
sns.set()
```

```python
def correlationCoefficientStartVSstop(day):
    df_m1=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\day{}\m1\__start_stop_compare.csv'.format(day))
    df_m2=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\day{}\m2\__start_stop_compare.csv'.format(day))
    df_e1=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\day{}\e1\__start_stop_compare.csv'.format(day))
    df_e2=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\day{}\e2\__start_stop_compare.csv'.format(day))
    df_n1=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\day{}\n1\__start_stop_compare.csv'.format(day))
    df_n2=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\day{}\n2\__start_stop_compare.csv'.format(day))
   
    fig, ax = plt.subplots(3,2, figsize=(20,15))
    fig.suptitle('Correlation coefficients between Trip start counts and Trip End counts - 7 Aug')
    fig.subplots_adjust(hspace=0.4, top=0.85)
    #M1
    r_m1="{:.2f}".format(df_m1['starts_count'].corr(df_m1['ends_count'])) # calculate correlation coefficient and represent in 2 decimals
    m1_title='03:00-06:59   (r = '+str(r_m1)+' )' # create title by concating r value
    ax[0,0].set_title(m1_title, fontsize=18) #set title
    sns.regplot(ax=ax[0,0],x=df_m1['starts_count'], y=df_m1['ends_count'],ci=None) #plot scatter plot

    #M2
    r_m2="{:.2f}".format(df_m2['starts_count'].corr(df_m2['ends_count']))
    m2_title='07:00-11:59   (r = '+str(r_m2)+' )'
    ax[0,1].set_title(m2_title, fontsize=18)
    sns.regplot(ax=ax[0,1],x=df_m2['starts_count'], y=df_m2['ends_count'],ci=None)

    #E1
    r_e1="{:.2f}".format(df_e1['starts_count'].corr(df_e1['ends_count']))
    e1_title='12:000-15:59  (r = '+str(r_e1)+' )'
    ax[1,0].set_title(e1_title, fontsize=18)
    sns.regplot(ax=ax[1,0],x=df_e1['starts_count'], y=df_e1['ends_count'],ci=None)

    #E2
    r_e2="{:.2f}".format(df_e2['starts_count'].corr(df_e2['ends_count']))
    e2_title='16:000-18:59  (r = '+str(r_e2)+' )'
    ax[1,1].set_title(e2_title, fontsize=18)
    sns.regplot(ax=ax[1,1],x=df_e2['starts_count'], y=df_e2['ends_count'],ci=None)

    #N1
    r_n1="{:.2f}".format(df_n1['starts_count'].corr(df_n1['ends_count']))
    n1_title='10:000-21:59  (r = '+str(r_n1)+' )'
    ax[2,0].set_title(n1_title, fontsize=18)
    sns.regplot(ax=ax[2,0],x=df_n1['starts_count'], y=df_n1['ends_count'],ci=None)

    #N2
    r_n2="{:.2f}".format(df_n2['starts_count'].corr(df_n2['ends_count']))
    n2_title='22:000-02:59 (r = '+str(r_n2)+' )'
    ax[2,1].set_title(n2_title, fontsize=18)
    sns.regplot(ax=ax[2,1],x=df_n2['starts_count'], y=df_n2['ends_count'],ci=None)
```
