```python
import pandas as pd
```

```python
def createTimeSepStarts(day):
    #3,7 am
    morning1=[]
    for hour in range(3,7):
        try:
            df=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\August\Started\Hourly\{}_{}.csv'.format(day,hour))
        except FileNotFoundError:
            continue
        morning1.append(df)
    df_m1 = pd.concat(morning1, axis=0).reset_index()
    df_m1.to_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\{}day_m1.csv'.format(day,hour),index=False)
    #7-12 am
    morning2=[]
    for hour in range(7,12):
        try:
            df=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\August\Started\Hourly\{}_{}.csv'.format(day,hour))
        except FileNotFoundError:
            continue
        morning2.append(df)
    df_m2 = pd.concat(morning2, axis=0).reset_index()
    df_m2.to_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\{}day_m2.csv'.format(day,hour),index=False)
    #12-4pm
    evening1=[]
    for hour in range(12,16):
        try:
            df=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\August\Started\Hourly\{}_{}.csv'.format(day,hour))
        except FileNotFoundError:
            continue
        evening1.append(df)
    df_e1 = pd.concat(evening1, axis=0).reset_index()
    df_e1.to_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\{}day_e1.csv'.format(day,hour),index=False)
    #4-7pm
    evening2=[]
    for hour in range(16,19):
        try:
            df=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\August\Started\Hourly\{}_{}.csv'.format(day,hour))
        except FileNotFoundError:
            continue
        evening2.append(df)
    df_e2 = pd.concat(evening2, axis=0).reset_index()
    df_e2.to_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\{}day_e2.csv'.format(day,hour),index=False)
    #7-10pm
    nigh1=[]
    for hour in range(19,22):
        try:
            df=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\August\Started\Hourly\{}_{}.csv'.format(day,hour))
        except FileNotFoundError:
            continue
        nigh1.append(df)
    df_n1 = pd.concat(nigh1, axis=0).reset_index()
    df_n1.to_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\{}day_n1.csv'.format(day,hour),index=False)
    #10-3am
    night2_1=[]
    night2_2=[]
    for hour in range(22,24):
        try:
            df1=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\August\Started\Hourly\{}_{}.csv'.format(day,hour))
        except FileNotFoundError:
            continue
        night2_1.append(df1)
    for hour in range(1,3):
        try:
            df2=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\August\Started\Hourly\{}_{}.csv'.format(day+1,hour))
        except FileNotFoundError:
            continue
        night2_2.append(df2)
    df=night2_1+night2_2
    df_n2 = pd.concat(df, axis=0).reset_index()
    df_n2.to_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\{}day_n2.csv'.format(day,hour),index=False)
```
