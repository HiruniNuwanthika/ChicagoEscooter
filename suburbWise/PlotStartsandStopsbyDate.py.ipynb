```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_stop=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\stop\day1\e1\__suburbWiseCount.csv')
```

```python
df_start=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\day1\e1\__suburbWiseCount.csv')
```

```python
df_stop.head()
```

```python
df_start.head()
```

```python
df_both=df_start.merge(df_stop, on="suburb", how="outer")
```

```python
df_both.columns=["suburb","starts_count", "ends_count"]
```

```python
df_both.to_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\day1\e1\__start_stop_compare.csv')
```

```python
barWidth = 0.25
fig = plt.figure(figsize = (20,15))

sub=df_both['suburb']
start=df_both['starts_count']
stop=df_both['ends_count']

br1 = np.arange(len(start))
br2 = [x + barWidth for x in br1]

plt.bar(br1, start, color ='r', width = barWidth,
        edgecolor ='grey', label ='Trip Starts')
plt.bar(br2, stop, color ='g', width = barWidth,
        edgecolor ='grey', label ='Trip Stops')
plt.xlabel('Suburbs', fontweight ='bold', fontsize = 15)
plt.ylabel('Number of Trips', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(start))],sub,rotation=90)
plt.legend()
plt.show()
```

```python
part_list=['m1','m2','e1','e2','n1','n2']
for part in part_list:
    df_stop=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\stop\day1\{}\__suburbWiseCount.csv'.format(part))
    df_start=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\day1\{}\__suburbWiseCount.csv'.format(part))
    df_both=df_start.merge(df_stop, on="suburb", how="outer")
    df_both.columns=["suburb","starts_count", "ends_count"]
    df_both.to_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\day1\{}\__start_stop_compare.csv'.format(part),index=False)
```

```python
df_m1=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\day1\m1\__start_stop_compare.csv')
df_m2=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\day1\m2\__start_stop_compare.csv')
df_e1=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\day1\e1\__start_stop_compare.csv')
df_e2=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\day1\e2\__start_stop_compare.csv')
df_n1=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\day1\n1\__start_stop_compare.csv')
df_n2=pd.read_csv(r'C:\Melbourne_Escooter\RealDataset\FinalResults\SuburbWise\August\start\day1\n2\__start_stop_compare.csv')
```

```python
#x, y1, y2 axis values taken from DFs
sub_m1=df_m1['suburb'] #x
start_m1=df_m1['starts_count'] #y1
stop_m1=df_m1['ends_count'] #y2
sub_m2=df_m2['suburb']
start_m2=df_m2['starts_count']
stop_m2=df_m2['ends_count']
sub_e1=df_e1['suburb']
start_e1=df_e1['starts_count']
stop_e1=df_e1['ends_count']
sub_e2=df_e2['suburb']
start_e2=df_e2['starts_count']
stop_e2=df_e2['ends_count']
sub_n1=df_n1['suburb']
start_n1=df_n1['starts_count']
stop_n1=df_n1['ends_count']
sub_n2=df_n2['suburb']
start_n2=df_n2['starts_count']
stop_n2=df_n2['ends_count']

#positions of x axis for all 6 subplots
x_m1 = np.arange(len(sub_m1))
x_m2 = np.arange(len(sub_m2))
x_e1 = np.arange(len(sub_e1))
x_e2 = np.arange(len(sub_e2))
x_n1 = np.arange(len(sub_n1))
x_n2 = np.arange(len(sub_n2))
width = 0.3

fig=plt.figure(figsize=(20,15))

fig.subplots_adjust(hspace=0.4, top=0.85)

fig.suptitle("1 August (Monday)", fontsize=20)

plt.subplot(321) #num_rows, num_columns, index
plt.title('Morning 1', fontsize=18)
plt.bar(x_m1-width/2, start_m1, width, label='starts', color ='r') #x,y
plt.bar(x_m1+width/2, stop_m1, width, label='stops', color ='g')
plt.xticks([r + width for r in range(len(sub_m1))],sub_m1,rotation=90)

plt.subplot(322)
plt.title('Morning 2', fontsize=18)
plt.bar(x_m2-width/2, start_m2, width, color ='r')
plt.bar(x_m2+width/2, stop_m2, width, color ='g')
plt.xticks([r + width for r in range(len(sub_m2))],sub_m2,rotation=90)

plt.subplot(323)
plt.title('Eve 1', fontsize=18)
plt.bar(x_e1-width/2, start_e1, width, color ='r')
plt.bar(x_e1+width/2, stop_e1, width, color ='g')
plt.xticks([r + width for r in range(len(sub_e1))],sub_e1,rotation=90)

plt.subplot(324)
plt.title('Eve 2', fontsize=18)
plt.bar(x_e2-width/2, start_e2, width, color ='r')
plt.bar(x_e2+width/2, stop_e2, width, color ='g')
plt.xticks([r + width for r in range(len(sub_e2))],sub_e2,rotation=90)

plt.subplot(325)
plt.title('Nig 1', fontsize=18)
plt.bar(x_n1-width/2, start_n1, width, color ='r')
plt.bar(x_n1+width/2, stop_n1, width, color ='g')
plt.xticks([r + width for r in range(len(sub_n1))],sub_n1,rotation=90)

plt.subplot(326)
plt.title('Nig 2', fontsize=18)
plt.bar(x_n2-width/2, start_n2, width, color ='r')
plt.bar(x_n2+width/2, stop_n2, width, color ='g')
plt.xticks([r + width for r in range(len(sub_n2))],sub_n2,rotation=90)

plt.figlegend(loc='upper right', ncol=1, labelspacing=0.5, fontsize=14, bbox_to_anchor=(1.11, 0.9))
plt.tight_layout(w_pad=6)
plt.show()

'''
plt.bar(br1, y1, color ='r', width = barWidth,
        edgecolor ='grey', label ='Trip Starts')
plt.bar(br2, y2, color ='g', width = barWidth,
        edgecolor ='grey', label ='Trip Stops')
plt.xlabel('Suburbs', fontweight ='bold', fontsize = 15)
plt.ylabel('Number of Trips', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(start))],sub,rotation=90)
plt.legend()
plt.show()
'''
```
